```python
#3. Write a Python program to count the number 4 in a given list.
#List = [1,4,6,8,4,9,4]

def list_count_4(nums):
  count = 0  
  for num in nums:
    if num == 4:
      count = count + 1

  return count

print(list_count_4([1,4,6,8,4,9,4]))
```

```python
#9. Write a Python function that takes a list and returns a new list with unique elements of the first list.
#Sample List : [1,2,3,3,3,3,4,5]

def unique_list(l):
  x = []
  for a in l:
    if a not in x:
      x.append(a)
  return x

print(unique_list([1,2,3,3,3,3,4,5]))
```

```python
#1. Write a Python program to concatenate all elements in a list into a string and return it.

def concatenate_list_data(list):
    result= ''
    for element in list:
        result += str(element)
    return result

print(concatenate_list_data([14, 32, 13, 51]))
```

```python
#1. Write Python Programs to use various operators in Python

#Arithmetic Operators
#Lets assign values to our variables
x = 20
y = 5

#Addition Operator
print('x + y =',x+y)

#Subtraction Operator
print('x - y =',x-y)

#Multiplication Operator
print('x * y =',x*y)

#Division Operator
print('x / y =',x/y)

#Percentail Operator
print ('x % y =', x % y)

#Floor Division OPerator
print('x // y =',x//y)

#Exponent Operator
print('x ** y =',x**y)
```

```python
#Comparision Operators

#Lets assign values to our variables
x = 15
y = 30

# Greater Than Operator
print('x > y is',x>y)

# Less Than Operator
print('x < y is',x<y)

# Equal Comparison Operator
print('x == y is',x==y)

# Not Equal Comparison Operator
print('x != y is',x!=y)

# Greater Than or Equal to operator
print('x >= y is',x>=y)

# Less than or equal to operator
print('x <= y is',x<=y)
```

```python
#Assignment Operators

#Assign value Operator
#This operator assigns values to the right hand operand from the left side operand.
print ('3 + 4 =', 3 + 4)
```

```python
#logical Operators

#Lets assign values to our variables
x = -50
y = 100

#Logical AND operator
print('x and y is',x and y)

#Logical OR operator
print('x or y is',x or y)

#Logical NOT
print('not x is',not x)
```

```python
#Bitwise Operators

#Lets assign values to our variables
x = 333
y = 111

#AND Bitwise Operator
#This operator performs bit by bit AND operations on the values on either side of the operator.
print ('x & y =', x & y)

#OR Bitwise Operator
#This operator performs bit by bit OR operations on the values on either side of the operator.
print ('x | y =', x | y)

#XOR Bitwise Operator
#This operator performs bit by bit exclusive OR operations on the values on either side of the operator.
print ('x ^ y =', x ^ y)

#Ones Complement Bitwise Operator
#The ones complement of a numbers binary is returned.
print ('~x =', ~x)

#Left shift Bitwise Operator
#The value to the left operator is shifted to the left as many times as the value on the right side of the operator.
print('x << 3 =', x << 3)

#Right shift Bitwise Operator
#The value to the left of the operator is shifted to the right as many times as the value on the right of the operator.
print('x >> 3 =', x >> 3)
```

```python
#Membership operators

#In Membership Operator
#This In operator is used in python to evaluate if a particular value is there or not in a sequence.
#The In operator returns True if it the specified element found in the list, otherwise it returns false.

#Lets assign list to our variables
x  = (1,2,3,4,5,6,7,8,9)
y  = (10,11,12,13,14,15)

print ('5 is in x =', 5 in x)

print ('5 is in y =', 5 in y)

#Not In Membership Operator
#Not in operator, the name itself is expressing the meaning that something is not inside.
#If it finds there is no such value, it returns True, otherwise False.

#Lets assign list to our variables
x  = (1,2,3,4,5,6,7,8,9)
y  = (10,11,12,13,14,15)

print ('5 is not in x =', 5 not in x)

print ('5 is not in y =', 5 not in y)
```

```python
#Identity operators

#This operator returns true statement if two identical operands reside in the same memory, otherwise false.
x1 = 5
y1 = 5
x2 = 'Hello'
y2 = 'Hello'
x3 = [1,2,3]
y3 = [1,2,3]

print('x1 is y1 = ', x1 is y1)

print('x2 is y2 =', x2 is y2)

print('x3 is y3 = ', x3 is y3)

#Is not Identity operator
#It will return false if either side of the operator pointing towards the same thing, otherwise it returns true.

print('x1 is not y1 = ', x1 is not y1)

print('x2 is not y2 =', x2 is not y2)

print('x3 is not y3 = ', x3 is not y3)
```

```python
#4. Write a python program to Read & write Excel files 

from openpyxl import Workbook

book = Workbook()
sheet = book.active

sheet['A3'] = 39
sheet['B3'] = 19

rows = [
    (88, 46),
    (89, 38),
    (23, 59),
    (56, 21),
    (24, 18),
    (34, 15)
]

for row in rows:
    sheet.append(row)

print(sheet.dimensions)
print("Minimum row: {0}".format(sheet.min_row))
print("Maximum row: {0}".format(sheet.max_row))
print("Minimum column: {0}".format(sheet.min_column))
print("Maximum column: {0}".format(sheet.max_column))

for c1, c2 in sheet[sheet.dimensions]:
    print(c1.value, c2.value)

book.save('dimensions.xlsx')
```

```python
#5. Write a python program to scrape reviews from a commercial web site

#import packages
from bs4 import BeautifulSoup
import urllib.request

#the targeted URL
url = 'https://www.yelp.com/biz/milk-and-cream-cereal-bar-new-york?osq=Ice+Cream'

#use request to open the URL
ourUrl = urllib.request.urlopen(url)

#create a BeautifulSoup object, which represents the document as a nested data structure
soup = BeautifulSoup(ourUrl, 'html.parser')

#to see what inside the soup
print (soup.prettify())

review = [] #create an empty list to store reviews
for i in soup.find_all('div', {'class' : 'review-content'}):
    per_review=i.find('p') #extract review
    print(per_review)
    review.append(per_review) #append review
    
len(review) #how many reviews we collect

#basic clean
New_review=[] #create an empty list to store new reviews
for each in review:
    new_each=str(each).replace('<br/>', '') #remove <br/> with empty, which means delete
    new_each=new_each[13:-4] #remove first 13 strings and last 4 strings, which are <p lang=' en'> and </p>
    print (new_each)
    New_review.append(new_each)
```
